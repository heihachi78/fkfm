```python
import cv2
import numpy as np
#from google.colab import files
#from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

#uploaded = files.upload()
#img_path = list(uploaded.keys())[0]
image = cv2.imread("book2.jpg")
print(image.shape)
#cv2_imshow(image)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

Globális átméretezés beállítása.

```python
scaling_factor = 4

def set_global_figsize():
    plt.rcParams['figure.figsize'] = [val * scaling_factor for val in plt.rcParams['figure.figsize']]

# Call the function once to initialize the global figure size
set_global_figsize()
```

Futtassuk a Canny algoritmust 3 különböző beállítással.

```python
edges1 = cv2.Canny(image, 100, 200)  # alsó küszöb: 100, felső küszöb: 200
edges2 = cv2.Canny(image, 150, 250)   #
edges3 = cv2.Canny(image, 50, 150)  #

# Display the results
#plt.figure(figsize=(10,5))
plt.subplot(131), plt.imshow(edges1, cmap='gray')
plt.title('Edges with thresholds 100, 200'), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(edges2, cmap='gray')
plt.title('Edges with thresholds 150, 250'), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(edges3, cmap='gray')
plt.title('Edges with thresholds 50, 150'), plt.xticks([]), plt.yticks([])
```
